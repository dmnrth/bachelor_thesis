```python
#functions needed for optimization

def portfolio_r(weights):
    return(np.dot(weights, mean_returns))

def portfolio_sd(weights):
    return(np.sqrt(np.dot(np.transpose(weights), np.dot(cov_matrix, weights))))

def portfolio_var(weights):
    return(np.dot(np.transpose(weights), np.dot(cov_matrix, weights)))

def information_ratio(weights):
    return(- (portfolio_r(weights))/portfolio_sd(weights))
```

```python
#functions needed for time data management

def get_month_length(timestamp):
    year = timestamp.year
    if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
        month_days_number = [31,29,31,30,31,30,31,31,30,31,30,31]
    else:
        month_days_number = [31,28,31,30,31,30,31,31,30,31,30,31]
    return(month_days_number[timestamp.month - 1])

def months_forward(timestamp, n):
    if(n == 0):
        return(timestamp)
    
    new_timestamp = timestamp
    for i in range (0, n):
        new_timestamp = new_timestamp + pd.Timedelta(get_month_length(new_timestamp), 'd')
    return new_timestamp

#function to convert monthly weights to daily weights (copy monthly wages to every day of the month)

def convert_monthly_to_daily(dataframe):
    daily_df = price_data.copy()

    daily_df[:] = 0

    daily_df = daily_df.loc["2007-01-01":]

    for i in range(0, len(daily_df)):

        d_year = daily_df.index[i].year
        d_month = daily_df.index[i].month
        date_str = str(d_year) + "-" + str(d_month) + "-" + str(1)
        datestamp = pd.to_datetime(date_str, format = "%Y-%m-%d")
        daily_df.iloc[i] = dataframe.loc[datestamp.strftime(format = "%Y-%m-%d")]

    return daily_df
```

# Equity line calculation for MV

### Global maximum information ratio

```python
#n - number of stocks
#f - rebalancing period length
#t - estimation window length
#c - transactional costs
#est_date - start date of estimation period

def get_mv_gmir_equity_line(str_name, start_date, est_date, n, f, t, c):

    wts = pd.DataFrame()
    
    for i in range(0, 192):

        est_start = months_forward(est_date, i)
        est_end = months_forward(est_date, i + 12) - pd.Timedelta(1, unit = 'day')

        date = months_forward(start_date, i)

        est_data = daily_returns2.loc[est_start:est_end]

        est_data = est_data[df.loc[date]]

        if(date <= pd.to_datetime("2009-06-01")):
            est_data = est_data.loc[:, est_data.columns != "GM"]

        est_data = est_data.sort_index(axis = 1)

        global mean_returns
        mean_returns = est_data.mean(axis = 0) * 252

        global cov_matrix
        cov_matrix = np.cov(est_data.transpose()) * 252

        if(date.strftime("%Y-%m-%d") <= "2009-06-01"):
            x0 = np.full(29, 1/29)
            bounds = tuple((0,1) for i in range(0, 29))
        else:
            x0 = np.full(30, 1/30)
            bounds = tuple((0,1) for i in range(0, 30))

        results = sp.optimize.minimize(
        fun = information_ratio,
        x0 = x0,
        method = 'SLSQP',
        bounds = bounds,
        constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1}))

        results["x"]

        results_normalized = results["x"] / sum(results["x"])

        index_names = df.loc[date]
        index_names = index_names[index_names != "GM"]

        results_series = pd.Series(results_normalized, index = index_names, name = date)

        if(date.strftime("%Y-%m-%d") <= "2009-06-01"):
            results_series["GM"] = 0

        wts = pd.concat((wts, results_series), axis = 1)

    wts = wts.transpose()
    wts = wts.fillna(0)
    wts = wts.sort_index(axis = 1)

    #number of stocks constituting at least 75%, 90% of the portfolio

    n75 = []
    n90 = []

    for x in range(0, len(wts)):
        temp_array = list(wts.iloc[x].copy())

        wsum = 0
        treshold = 0.75
        counter = 0

        while np.round(wsum, decimals = 2) < treshold:
            wsum = wsum + max(temp_array)
            counter = counter + 1
            temp_array.remove(max(temp_array))

        n75.append(counter)
    
    for x in range(0, len(wts)):
        temp_array = list(wts.iloc[x].copy())

        wsum = 0
        treshold = 0.90
        counter = 0

        while np.round(wsum, decimals = 2) < treshold:
            wsum = wsum + max(temp_array)
            counter = counter + 1
            temp_array.remove(max(temp_array))

        n90.append(counter)
    
    perf_stats_pt1[str_name] = [sum(n75)/len(n75), sum(n90)/len(n90)]

    # #daily weights

    dwts = convert_monthly_to_daily(wts)

    #daily returns (+1)

    dret = np.multiply(dwts, daily_returns)
    dret = dret.sum(axis = 1)
    dret[:] = dret[:] + 1

    #transactional costs

    trans_costs = dwts - dwts.shift(1)
    trans_costs.iloc[0] = dwts.iloc[0]
    trans_costs = np.abs(trans_costs)
    trans_costs = trans_costs.sum(axis = 1)
    trans_costs = trans_costs * c

    dret = dret - trans_costs

    #equity line

    eqline = pd.Series(np.ones(len(dret)))
    for i in range(0, len(eqline)):
        eqline[i + 1] = eqline[i] * dret[i]

    #editing indices

    ind = dret.index.tolist()
    ind.insert(0, pd.to_datetime("2006-12-29"))

    eqline.index = ind

    return(eqline)
```

```python
#n - number of stocks
#f - rebalancing period length
#t - estimation window length
#c - transactional costs

def get_fmv_gmir_equity_line(str_name, path, start_date, n, f, t, c):

    wts = pd.DataFrame()
    
    for i in range(0, 192):

        date = months_forward(start_date, i)

        forecasts = pd.read_csv(path + date.strftime("%Y-%m-%d") + ".csv")
        forecasts = forecasts.iloc[:,1:]
        forecasts = forecasts.sort_index(axis = 1)

        #DOW error fix
        if(forecasts.iloc[1:20,1:20].isnull().sum().sum() != 0):
            forecasts["DOW"].iloc[len(forecasts["DOW"])-30:len(forecasts["DOW"])] = forecasts["DOW"].iloc[0:30]
            forecasts = forecasts.iloc[len(forecasts["DOW"])-30:len(forecasts["DOW"])]
            if(forecasts.iloc[1:20,1:20].isnull().sum().sum() != 0): print("Still wrong")

        global mean_returns 
        mean_returns = forecasts.mean(axis = 0) * 252
        global cov_matrix
        cov_matrix = np.cov(forecasts.transpose()) * 252

        if(date.strftime("%Y-%m-%d") <= "2009-06-01"):
            x0 = np.full(29, 1/29)
            bounds = tuple((0,1) for i in range(0, 29))
        else:
            x0 = np.full(30, 1/30)
            bounds = tuple((0,1) for i in range(0, 30))

        results = sp.optimize.minimize(
        fun = information_ratio,
        x0 = x0,
        method = 'SLSQP',
        bounds = bounds,
        constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1}))

        results["x"]

        results_normalized = results["x"] / sum(results["x"])

        results_series = pd.Series(results_normalized, index = forecasts.columns, name = date)

        if(date.strftime("%Y-%m-%d") <= "2009-06-01"):
            results_series["GM"] = 0

        wts = pd.concat((wts, results_series), axis = 1)

    wts = wts.transpose()
    wts = wts.fillna(0)
    wts = wts.sort_index(axis = 1)

    n75 = []
    n90 = []

    #number of stocks constituting at least 75%, 90% of the portfolio

    for x in range(0, len(wts)):
        temp_array = list(wts.iloc[x].copy())

        wsum = 0
        treshold = 0.75
        counter = 0

        while np.round(wsum, decimals = 2) < treshold:
            wsum = wsum + max(temp_array)
            counter = counter + 1
            temp_array.remove(max(temp_array))

        n75.append(counter)
    
    for x in range(0, len(wts)):
        temp_array = list(wts.iloc[x].copy())

        wsum = 0
        treshold = 0.90
        counter = 0

        while np.round(wsum, decimals = 2) < treshold:
            wsum = wsum + max(temp_array)
            counter = counter + 1
            temp_array.remove(max(temp_array))

        n90.append(counter)
    
    perf_stats_pt1[str_name] = [sum(n75)/len(n75), sum(n90)/len(n90)]

    #daily weights

    dwts = convert_monthly_to_daily(wts)

    #daily returns (+1)

    dret = np.multiply(dwts, daily_returns)
    dret = dret.sum(axis = 1)
    dret[:] = dret[:] + 1

    #transactional costs

    trans_costs = dwts - dwts.shift(1)
    trans_costs.iloc[0] = dwts.iloc[0]
    trans_costs = np.abs(trans_costs)
    trans_costs = trans_costs.sum(axis = 1)
    trans_costs = trans_costs * c

    dret = dret - trans_costs

    #equity line

    eqline = pd.Series(np.ones(len(dret)))
    for i in range(0, len(eqline)):
        eqline[i + 1] = eqline[i] * dret[i]

    #editing indices

    ind = dret.index.tolist()
    ind.insert(0, pd.to_datetime("2006-12-29"))

    eqline.index = ind

    return(eqline)
```

## SA: ARIMA-GARCH HML

```python
#n - number of stocks
#f - rebalancing period length
#t - estimation window length
#c - transactional costs

def get_fmv_gmir_equity_line_HSL(str_name, path, start_date, n, f, t, c, GM_corr):

    wts = pd.DataFrame()
    
    for i in range(0, 192):

        date = months_forward(start_date, i)

        forecasts = pd.read_csv(path + date.strftime("%Y-%m-%d") + ".csv")
        forecasts = forecasts.iloc[:,1:]
        forecasts = forecasts.sort_index(axis = 1)

        #DOW error fix
        if(forecasts.iloc[1:20,1:20].isnull().sum().sum() != 0):
            forecasts["DOW"].iloc[len(forecasts["DOW"])-30:len(forecasts["DOW"])] = forecasts["DOW"].iloc[0:30]
            forecasts = forecasts.iloc[len(forecasts["DOW"])-30:len(forecasts["DOW"])]
            if(forecasts.iloc[1:20,1:20].isnull().sum().sum() != 0): print("Still wrong")

        global mean_returns 
        mean_returns = forecasts.mean(axis = 0) * 252
        global cov_matrix
        cov_matrix = np.cov(forecasts.transpose()) * 252

        if(date.strftime("%Y-%m-%d") <= "2009-06-01" and GM_corr == True):
            x0 = np.full(9, 1/9)
            bounds = tuple((0,1) for i in range(0, 9))
        else:
            x0 = np.full(10, 1/10)
            bounds = tuple((0,1) for i in range(0, 10))

        results = sp.optimize.minimize(
        fun = information_ratio,
        x0 = x0,
        method = 'SLSQP',
        bounds = bounds,
        constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1}))

        results["x"]

        results_normalized = results["x"] / sum(results["x"])

        results_series = pd.Series(results_normalized, index = forecasts.columns, name = date)

        if(date.strftime("%Y-%m-%d") <= "2009-06-01" and GM_corr == True):
            results_series["GM"] = 0

        wts = pd.concat((wts, results_series), axis = 1)

    wts = wts.transpose()
    wts = wts.fillna(0)
    wts = wts.sort_index(axis = 1)

    n75 = []
    n90 = []

    #number of stocks constituting at least 75%, 90% of the portfolio

    for x in range(0, len(wts)):
        temp_array = list(wts.iloc[x].copy())

        wsum = 0
        treshold = 0.75
        counter = 0

        while np.round(wsum, decimals = 2) < treshold:
            wsum = wsum + max(temp_array)
            counter = counter + 1
            temp_array.remove(max(temp_array))

        n75.append(counter)
    
    for x in range(0, len(wts)):
        temp_array = list(wts.iloc[x].copy())

        wsum = 0
        treshold = 0.90
        counter = 0

        while np.round(wsum, decimals = 2) < treshold:
            wsum = wsum + max(temp_array)
            counter = counter + 1
            temp_array.remove(max(temp_array))

        n90.append(counter)
    
    perf_stats_pt1[str_name] = [sum(n75)/len(n75), sum(n90)/len(n90)]
    
    all_tickers = list(price_data.columns)
    these_tickers = list(wts.columns)
    other_tickers = []

    for x in all_tickers:
        if x not in these_tickers:
            other_tickers.append(x)

    for x in other_tickers:
        wts[x] = np.zeros(192)

    wts = wts.sort_index(axis = 1)

    #daily weights

    dwts = convert_monthly_to_daily(wts)

    #daily returns (+1)

    dret = np.multiply(dwts, daily_returns)
    dret = dret.sum(axis = 1)
    dret[:] = dret[:] + 1

    #transactional costs

    trans_costs = dwts - dwts.shift(1)
    trans_costs.iloc[0] = dwts.iloc[0]
    trans_costs = np.abs(trans_costs)
    trans_costs = trans_costs.sum(axis = 1)
    trans_costs = trans_costs * c

    dret = dret - trans_costs

    #equity line

    eqline = pd.Series(np.ones(len(dret)))
    for i in range(0, len(eqline)):
        eqline[i + 1] = eqline[i] * dret[i]

    #editing indices

    ind = dret.index.tolist()
    ind.insert(0, pd.to_datetime("2006-12-29"))

    eqline.index = ind

    return(eqline)
```
